```python
import matplotlib.pyplot as plt

# Data
harris_total_words = 91580
trump_total_words = 297170
number_of_speeches = 30

# Calculate average words per speech
harris_avg_words = harris_total_words / number_of_speeches
trump_avg_words = trump_total_words / number_of_speeches

# Data for visualization
speakers = ['Kamala Harris', 'Donald Trump']
average_words = [harris_avg_words, trump_avg_words]

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(speakers, average_words, color=['orange', 'blue'], alpha=0.7)

# Add labels and title
plt.xlabel('Speakers', fontsize=14)
plt.ylabel('Average Words per Speech', fontsize=14)
plt.title('Comparative Analysis of Average Words per Speech', fontsize=16)
plt.ylim(0, max(average_words) * 1.1)  # Add some space above the tallest bar

# Annotate the bars with values
for i, value in enumerate(average_words):
    plt.text(i, value + 500, f'{value:.2f}', ha='center', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
