```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Harris
harris_data = {
    "interrogative": 10,
    "imperative": 12,
    "declarative": 72,
    "exclamative": 5
}

# Data for Trump
trump_data = {
    "interrogative": 11,
    "imperative": 17,
    "declarative": 138,
    "exclamative": 6
}

# Calculate percentages for Harris
harris_total = sum(harris_data.values())
harris_percentages = {stype: (value / harris_total) * 100 for stype, value in harris_data.items()}

# Calculate percentages for Trump
trump_total = sum(trump_data.values())
trump_percentages = {stype: (value / trump_total) * 100 for stype, value in trump_data.items()}

# Sentence types
sentence_types = list(harris_percentages.keys())

# Values for each speaker
harris_values = [harris_percentages[stype] for stype in sentence_types]
trump_values = [trump_percentages[stype] for stype in sentence_types]

# Bar width
bar_width = 0.35

# Positions on the x-axis
x = np.arange(len(sentence_types))

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot bars
ax.bar(x - bar_width/2, harris_values, bar_width, label='Harris', color='orange', alpha=0.7)
ax.bar(x + bar_width/2, trump_values, bar_width, label='Trump', color='blue', alpha=0.7)

# Add labels, title, and legend
ax.set_xlabel('Sentence Types', fontsize=12)
ax.set_ylabel('Percentage (%)', fontsize=12)
ax.set_title('Comparative Analysis of Sentence Types (Percentage)', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(sentence_types, fontsize=10)
ax.legend(fontsize=10)

# Show grid for better readability
ax.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```
